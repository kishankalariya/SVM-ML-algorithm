# Machine Learning Assignment 3

```python
#wine dataset
from sklearn.datasets import load_wine
wine = load_wine()
```

### task-2

```python
#split the detaset 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine['data'],wine['target'], random_state=3004)
```

### Task-3

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
import numpy as np
```

```python
svm = SVC()
np.mean(cross_val_score(svm, X_train,y_train))
```

### Task-4

```python
svm = SVC()
svm.fit(X_train, y_train)

print("Accuracy on training set: ", svm.score(X_train , y_train))
print("Accuracy on test set: ", svm.score(X_test, y_test))
print("test error rate: ", 1-svm.score(X_test, y_test))
```

Using default values of parameters of svm gives 31% test error rate , which is not significant ,so next I do normlization as well as hyperparameter tuning     

### task-5

cell 7 to 10 are just for analysis purpose

```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)



svm = SVC()
svm.fit(X_train_scaled, y_train)

print("Accuracy on training set:", svm.score(X_train_scaled, y_train))
print("Accuracy on test set:", svm.score(X_test_scaled, y_test))
```

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)



svm = SVC()
svm.fit(X_train_scaled, y_train)

print("Accuracy on training set:", svm.score(X_train_scaled, y_train))
print("Accuracy on test set:", svm.score(X_test_scaled, y_test))
```

```python
#Normalizer
from sklearn.preprocessing import Normalizer

scaler = Normalizer()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)



svm = SVC()
svm.fit(X_train_scaled, y_train)

print("Accuracy on training set:", svm.score(X_train_scaled, y_train))
print("Accuracy on test set:", svm.score(X_test_scaled, y_test))
```

```python
#RobustScaler
from sklearn.preprocessing import RobustScaler

scaler= RobustScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)



svm = SVC()
svm.fit(X_train_scaled, y_train)

print("Accuracy on training set:", svm.score(X_train_scaled, y_train))
print("Accuracy on test set:", svm.score(X_test_scaled, y_test))
```

 According to result minmaxscaler is best and worst seen in Normalizer.In fact, robustscaler and  StandardScaler also give good score ,however not better then the MinMaxscaler.

### Task-6

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV


pipe1 = make_pipeline(MinMaxScaler(), SVC())
pipe2 = make_pipeline(Normalizer(), SVC())
pipe3 = make_pipeline(RobustScaler(), SVC())
pipe4 = make_pipeline(StandardScaler(), SVC())
```

```python
parameter = {'svc__C': [0.001,0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100,200],
              'svc__gamma': [0.001,0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100,200]} 
```

```python
# for minmaxscaler
grid = GridSearchCV(pipe1,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

```python
#for Normalizer
grid = GridSearchCV(pipe2,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

```python
#RobustScaler
grid = GridSearchCV(pipe3,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

```python
# StandardScaler
grid = GridSearchCV(pipe4,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

Here,best result got from standardscaler and minmaxscaler , which is 0% test error rate , whereas at Normalizer observed 8% error rate.

```python
# MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

#MLPClassifier
mlp = MLPClassifier(max_iter=100,random_state=3004)
#parameter tuning
parameter_space = {
    'hidden_layer_sizes': [(10,30,10),(20,)],
    'activation': ['identity', 'logistic','tanh', 'relu'],
    'solver': ['sgd', 'adam','lbfgs'],
    'alpha': [0.0001,0.001,0.01,0.1,1],
    'learning_rate': ['constant','adaptive'],
}

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=5)
grid.fit(X_train_scaled, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test_scaled, y_test))
print("Test error rate:", 1-grid.score(X_test_scaled, y_test))
print("Best parameters:", grid.best_params_)
```

# All tasks for  USPC  dataset

```python
#USPC dataset
import pandas as pd
df1 = pd.read_csv('zip.train.gz', compression='gzip', sep=' ',header=None, quotechar='"', error_bad_lines=False)
df2=pd.read_csv('zip.test (3).gz', compression='gzip', sep=' ',header=None, quotechar='"', error_bad_lines=False)

df1 = df1.drop([257],axis=1)
df = pd.concat([df1,df2])

y=df[0]
X= df.drop([0],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=3004)
```

```python
#task3 & 4

svm = SVC()
print("Accuracy on training set using cross val score: ", np.mean(cross_val_score(svm, X_train,y_train)))

svm = SVC()
svm.fit(X_train, y_train)

print("Accuracy on training set: ", svm.score(X_train , y_train))
print("Accuracy on test set: ", svm.score(X_test, y_test))
print("test error rate: ", 1-svm.score(X_test, y_test))

```

here,svc score is significant because only 2% test error rate seen

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV


pipe1 = make_pipeline(MinMaxScaler(), SVC())
pipe2 = make_pipeline(Normalizer(), SVC())

parameter = {'svc__C': [0.001,0.01,0.1],
              'svc__gamma': [0.001,0.01,0.1]} 

# for minmaxscaler
grid = GridSearchCV(pipe1,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

```python
#for Normalizer
grid = GridSearchCV(pipe2,parameter, cv=5)
grid.fit(X_train, y_train)
print("Best cross-validation accuracy:", grid.best_score_)
print("Test set score:", grid.score(X_test, y_test))
print("Test error rate:", 1-grid.score(X_test, y_test))
print("Best parameters:", grid.best_params_)
```

Here,best result got from minmaxscaler , which is 6% test error rate , whereas at Normalizer observed 12% error rate.More both the case best parameter is different.
